```python
import googleapiclient.discovery
import pandas as pd

api_service_name = "youtube"
api_version = "v3"
DEVELOPER_KEY = "AIzaSyBdtZSvCy7KTUews2RDE0p4cfF-y-qzcYQ"

youtube = googleapiclient.discovery.build(api_service_name, api_version, developerKey=DEVELOPER_KEY)

def get_comments(video_id, max_results=1000):
    comments = []
    request = youtube.commentThreads().list(
        part="snippet",
        videoId="vVqhgfoFaDk",
        maxResults=1000
    )

    while request and len(comments) < max_results:
        response = request.execute()

        for item in response['items']:
            comment = item['snippet']['topLevelComment']['snippet']
            comments.append([
                comment['authorDisplayName'],
                comment['publishedAt'],
                comment['updatedAt'],
                comment['likeCount'],
                comment['textDisplay']
            ])

            # Stop if we have reached the desired number of comments
            if len(comments) >= max_results:
                break

        # Check if there's a next page token to continue
        if 'nextPageToken' in response:
            request = youtube.commentThreads().list(
                part="snippet",
                videoId=video_id,
                pageToken=response['nextPageToken'],
                maxResults=100
            )
        else:
            request = None

    return comments

comments = get_comments("WNrB1Q9Rry0", max_results=500)
df = pd.DataFrame(comments, columns=['author', 'published_at', 'updated_at', 'like_count', 'Comment'])
print(f"Number of rows: {df.shape[0]}")
df.head(10)
```

```python
#Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import os

# Import functions for data preprocessing & data preparation
from sklearn.preprocessing import LabelEncoder
from sklearn.utils import resample
from sklearn.feature_extraction.text import CountVectorizer
from nltk.sentiment.vader import SentimentIntensityAnalyzer

from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.stem import PorterStemmer, LancasterStemmer
from nltk.stem.snowball import SnowballStemmer
from nltk.corpus import stopwords
from nltk.corpus import wordnet
import string
from string import punctuation
import nltk
import re
```

```python
data = pd.read_csv('youtube_comments.csv')
data.columns
data1=data.drop(['author','published_at','updated_at','like_count'],axis=1)
data1
```

```python
nltk.download('vader_lexicon')
sentiments = SentimentIntensityAnalyzer()
data1["Positive"] = [sentiments.polarity_scores(i)["pos"] for i in data1["Comment"]]
data1["Negative"] = [sentiments.polarity_scores(i)["neg"] for i in data1["Comment"]]
data1["Neutral"] = [sentiments.polarity_scores(i)["neu"] for i in data1["Comment"]]
data1['Compound'] = [sentiments.polarity_scores(i)["compound"] for i in data1["Comment"]]
score = data1["Compound"].values
sentiment = []
for i in score:
    if i >= 0.05 :
        sentiment.append('Positive')
    elif i <= -0.05 :
        sentiment.append('Negative')
    else:
        sentiment.append('Neutral')
data1["Sentiment"] = sentiment
data1.head()
```

```python
data2=data1.drop(['Positive','Negative','Neutral','Compound'],axis=1)
data2.head()
```

```python
# Text processing function
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
import nltk
nltk.download('stopwords')
```

```python
stop_words = stopwords.words('english')
lzr = WordNetLemmatizer()
```

```python
def text_processing(text):
    text = text.lower()
    text = re.sub(r'\n', ' ', text)
    text = re.sub('[%s]' % re.escape(punctuation), "", text)
    text = re.sub("^a-zA-Z0-9$,.", "", text)
    text = re.sub(r'\s+', ' ', text, flags=re.I)
    text = re.sub(r'\W', ' ', text)
    text = ' '.join([word for word in word_tokenize(text) if word not in stop_words])
    text = ' '.join([lzr.lemmatize(word) for word in word_tokenize(text)])
    return text
```

```python
data_copy = data2.copy()
data_copy.comment = data_copy.comment.apply(lambda text: text_processing(text))

processed_data = pd.DataFrame({
    'Sentence': data_copy.comment,
    'Sentiment': data_copy['Sentiment']
})
```

```python
# Resampling to balance the dataset
df_positive = processed_data[processed_data['Sentiment'] == 'Positive']
df_negative = processed_data[processed_data['Sentiment'] == 'Negative']
df_neutral = processed_data[processed_data['Sentiment'] == 'Neutral']
```

```python
positive_upsampled = resample(df_positive, replace=True, n_samples=len(df_neutral), random_state=27)
negative_upsampled = resample(df_negative, replace=True, n_samples=len(df_neutral), random_state=27)

upsampled = pd.concat([df_neutral, positive_upsampled, negative_upsampled])
```

```python
# Display value counts to check balance
print(upsampled['Sentiment'].value_counts())
```

```python

# Final balanced dataset
upsampled.tail()
```

```python
corpus = []
for sentence in upsampled['Sentence']:
    corpus.append(sentence)
corpus[0:5]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=1500)
X = cv.fit_transform(corpus).toarray()
y = upsampled.iloc[:, -1].values
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

```python
import seaborn as sns
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
y_pred = classifier.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
nb_score = accuracy_score(y_test, y_pred)
print('accuracy',nb_score)
```
